```python
import pandas as pd 
import numpy as np 
```

```python
df=pd.read_csv("iris.csv") # this iris.csv is stored in my folder 
df
```

```python
df.info()
```

```python
from  sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
df.columns.tolist()
```

```python
X=df[["sepal_length",'sepal_width','petal_length','petal_width']]
X # Features 
```

```python
Y=df['species']
Y  # Response 
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=45,test_size=0.2)


```

```python
mymodel=DecisionTreeClassifier() ## here we have created the object of This class 


mymodel.fit(x_train,y_train) #using the object of Decision class ie mymodel
# we can call the methods of that class so here we have  called the method 
# fit to train the model 
# once it is ready it predict anything 
```

```python
y_pred=mymodel.predict(x_test) # y_pred is new variable 
y_pred.tolist()
print(y_test)
print(y_pred.tolist())
```

```python
modacc=accuracy_score(y_test,y_pred)
modacc*100
```

```python
flower=[[4.5,2.3,5.2,2.4]]

res=mymodel.predict(flower)
res
```
