```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import  Lasso ,Ridge
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("fueldata.csv")
```

```python
features=df[['Engine Size(L)',
 'Cylinders',
 'Fuel Consumption City (L/100 km)',
 'Fuel Consumption Hwy (L/100 km)',
 'Fuel Consumption Comb (L/100 km)',
 'Fuel Consumption Comb (mpg)']]

X=features   # features 
Y=df['CO2 Emissions(g/km)'] # response 
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=49)
```

```python
X_train
```

```python
scaler=StandardScaler() # I have created the object of Standard Scaler class to use its methods 

X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
model=Ridge(alpha=0.5)

model.fit(X_train,Y_train)

```

```python
predic=model.predict(X_test)
```

```python
predic
```

```python
from sklearn.metrics  import r2_score
```

```python
acc=r2_score(Y_test,predic)
acc*100
```
