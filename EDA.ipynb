```python
import numpy as np 
# Here we are having the marks of 10 students 
marks=[20,30,35,40,45,50,55,60,70,80]
  
#Quartiles- _____Q1 _______Q2 ______Q3 _______

marks.sort()
Q1=np.percentile(marks,25)
Q2=np.percentile(marks,50)
Q3=np.percentile(marks,75)
print("The Quartiles are as follows Q1 = ",Q1," Q2 = ",Q2," Q3 =",Q3)
```

### Step 1 always find Q2 
#### Median/ Q2 =  n+1/2 -> 10+1/2 => 5.5
                5th value - 45
                6th  value -50 
                Q2=45+50/2 => 47.5


                
### 2 - Q1 
         Lower half data= [20,30,35,40,45]
         median is 35 so 
         Q1 = 35
#### Q3 
         upper half - [50,55,60,70,80]
                median =60 => Q3 


#### Quartiles are  
##### Q1= 30 ,Q2= 47.5 , Q3= 60

### IQR - Inter Quartile Range  
#### IQr is used to find spread of middle 50% of data or to detect the outliers 

```python
df=[2,4,6,8,10,12,14,16]
Q1=np.quantile(df,0.25)
Q2=np.quantile(df,0.50)
Q3=np.quantile(df,0.75)
print(Q1,Q2,Q3)


# IQR= Q3-Q1  - formula to calculate the IQR 
Iqr=Q3-Q1
Iqr
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

#First data set 
data1=[48,49,50,51,52]
#Second Dataset 
data2=[20,35,50,65,80]

mean1=np.mean(data1)
mean2=np.mean(data2)

std1=np.std(data1)
std2=np.std(data2)

plt.figure(figsize=(10,5))

plt.plot(data1,[1]*len(data1),'bo',label="DataSet first")
plt.plot(data2,[2]*len(data2),'ro',label="DataSet Second")

plt.axvline(mean1,color="blue",linestyle="--")
plt.axvline(mean2,color="red",linestyle="--")

plt.yticks([1,2],["Close Data ","Spread Data "])

plt.title("Variance and Standard Deviation")
plt.xlabel("Values")
plt.legend()
plt.show()


```

```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd 
```

```python
poskew=np.array([5,6,7,8,8,9,9,10,10,10,11,11,12,15,30])
plt.figure(figsize=(7,4))

print(np.mean(poskew),"  ",np.median(poskew))
```

```python
plt.hist(poskew,bins=6,color="steelblue",edgecolor="black",density=True)
sns.kdeplot(poskew,color="darkblue",linewidth=2)

plt.axvline(np.mean(poskew),color="red",linestyle="--",label="Mean")
plt.axvline(np.median(poskew),color="blue",linestyle="--",label="Median")

plt.title("Positive Skewed Data Where The Mean > Median ")

plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()


```

```python
negskew=np.array([20,35,40,42,43,44,45,45,46,47,48,48,49,50,50])
mn=np.mean(negskew)

mdn=np.median(negskew)
print("Mean",mn," Median",mdn)
```

```python
plt.hist(negskew,bins=6,color="steelblue",edgecolor="black",density=True)
sns.kdeplot(negskew,color="darkblue",linewidth=2)

plt.axvline(mn,color="red",linestyle="--",label="Mean")
plt.axvline(mdn,color="blue",linestyle="--",label="Median")

plt.title("Negative  Skewed Data Where The Mean < Median ")

plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

#### Skew 
 #### Positive Skew -  Mean > Median 
 #### Negative Skew -  Mean < Median 
 #### Neutral Skew  -  Mean == Median
