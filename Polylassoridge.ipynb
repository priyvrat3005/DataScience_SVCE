```python
import pandas as pd 
import numpy as np 

df=pd.DataFrame({
    "Age":[25,30,35,45,60,54,65,62,40,50],
    "Exp":[2,5,10,17,22,20,25,21,15,19],
    "Salary":[40000,64000,83000,110000,20000,45000,60000,560000,48000,150000]
})
df
```

```python
mean=df["Age"].mean()
mean ## 46.6 
std=df["Age"].std()
data=df["Age"]-mean  ## we are finding distance from the mean 
data
std
```

```python
df["Age"]=data/std

df["Age"]
# -3 to 3 
```

```python
mean_val=df["Exp"].mean()
std_val=df["Exp"].std()

df["Exp"]=(df["Exp"]-mean_val)/std_val
df["Exp"]

df

# scaling formula is  Z=(x-mean/std)

# -3  to  +3
```

```python
df=pd.read_csv("fueldata.csv")
df=df.rename(columns={"Fuel Consumption City (L/100 km)":"A","Fuel Consumption Hwy (L/100 km)":"B",
                   "Fuel Consumption Comb (L/100 km)":"C","Fuel Consumption Comb (mpg)":"D"})
df
```

```python
from sklearn.linear_model import Ridge 
from sklearn.metrics import r2_score


df=pd.read_csv("fueldata.csv")

features=df[['Engine Size(L)',
 'Cylinders',
 'Fuel Consumption City (L/100 km)',
 'Fuel Consumption Hwy (L/100 km)',
 'Fuel Consumption Comb (L/100 km)',
 'Fuel Consumption Comb (mpg)']]

X=features
Y=df['CO2 Emissions(g/km)'] # response 



model=Ridge(alpha=100000) ## alpha is he penalty strength to shring the coef of the features 

model.fit(X,Y)
```

```python
for col , coef in zip(X.columns,model.coef_):
    print(col,coef)
```
