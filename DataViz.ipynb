```python
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
df={"Studs":["Rahul","Rohit","Hemanth","Rishita","Raga"],
    "Marks":[85,78,91,79,86],
    "Age":[20,21,23,22,28]}

df=pd.DataFrame(df)
df
```

### Creating the normal graph in Matplotlib

```python
plt.figure(figsize=(7,4),facecolor="blue") # it creates a canvas on which will show tha graph 
plt.bar(df["Studs"],df["Marks"])
plt.title("Data Viz")
plt.show()
```

### LineGraph 

```python
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
months=["jan","feb","mar","apr","may"]
sales=[20000,23000,16000,30000,10000]

plt.figure(figsize=(7,4),facecolor="white")

plt.plot(months,sales,marker="*",markersize=10,label="Sales Data ",lw=1,ls="--")
plt.title("Data Viz ")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True,alpha=0.2)
plt.show()

```

### Scatter Plots 

```python
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
df2={
    "StudyHours":[1,2,3,4,5,2.5,3.5,4.5],
    "MarksScored":[50,55,65,70,85,60,68,75]
}
df2=pd.DataFrame(df2)
df2
```

```python
plt.figure(figsize=(7,4),facecolor="white")

plt.scatter(df2["StudyHours"],df2["MarksScored"],s=50,c="red",marker='^',lw=2,alpha=0.7)


plt.title("Scatter Plot ")
plt.grid(True,alpha=0.8)
plt.show()
```

```python
plt.figure(figsize=(7,4),facecolor="white")
plt.plot(df2["StudyHours"],df2["MarksScored"],marker="*",markersize=10,label="Sales Data ",lw=1,ls="--")
plt.title("Data Viz ")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True,alpha=0.2)
plt.show()
```
