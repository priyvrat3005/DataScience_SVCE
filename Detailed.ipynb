```python
import numpy as np
import matplotlib.pyplot as plt

# Data
data1 = [48, 49, 50, 51, 52]
data2 = [20, 35, 50, 65, 80]

mean1 = np.mean(data1)
mean2 = np.mean(data2)

# Standard deviation
sd1 = np.std(data1)
sd2 = np.std(data2)
```

```python
plt.figure(figsize=(10,5))

plt.plot(data1, [1]*len(data1), 'bo', label='Dataset 1')
plt.plot(data2, [2]*len(data2), 'ro', label='Dataset 2')

# Mean lines
plt.axvline(mean1, color='blue', linestyle='--')
plt.axvline(mean2, color='red', linestyle='--')


# plt.yticks([1,2], ['Close Data', 'Spread Data'])

plt.title("Variance and Standard Deviation")
plt.xlabel("Values")
plt.legend()

plt.show()
```

```python
print("Dataset 1 SD =", sd1)
print("Dataset 2 SD =", sd2)
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.array([
    [40, 60],
    [50, 55],
    [60, 70],
    [70, 90]
])

maths = data[:,0]
science = data[:,1]

plt.figure(figsize=(6,6))

plt.scatter(maths, science)

plt.xlabel("Maths Marks")
plt.ylabel("Science Marks")

plt.title("2D Data Distribution")

plt.grid(True)

plt.show()
```
