### Sampling 

```python
import numpy as np 
import pandas as pd 
```

```python
np.random.seed(42)

df=pd.DataFrame({ "ID":range(1,1001),"Gender":np.random.choice(["Male","Female"],size=1000,p=[0.7,0.3]
    ),
    "Marks":np.random.randint(40,100,1000)
})
df.head()
```

### Simple Random Sampling (SRS)
in this sampling every students have equal chance to be selected because we are selecting randomly 



```python
srs=df.sample(n=100,random_state=42)

srs.shape
```

```python
print(df["Marks"].mean()) ### Origin large data 
print(srs["Marks"].mean()) ### sample data taken from the original data 
```

### Startified Sampling method 
Making the groups before taking the sample data 
 then after creating the groups then  slecting the sample data from the groups and combine them to use  

 ##### These groups are called Strata 

```python
print(df["Gender"].value_counts())
```

```python
male_data=df[df["Gender"]=="Male"].sample(n=50,random_state=45)
fem_data=df[df["Gender"]=="Female"].sample(n=50,random_state=45)

# Here we have created Two startas male_data and fem_data

# Now we are going to add both the data together to get the final sample data 

final_data=pd.concat([male_data,fem_data])

final_data.shape
```
